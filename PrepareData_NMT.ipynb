```python
df = pd.read_excel('tmx_to_xlxs.xlsx', header=None)
```

```python
df.rename(columns={0: "tibetan_text", 1: "english_text"}, inplace= True)
df.head()
```

```python
X = df['english_text']
y = df['tibetan_text']
```

```python
from sklearn.model_selection import train_test_split

X_train_en, X_split_again_en, y_train_po, y_split_again_po = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_test_en, X_valid_en, y_test_po, y_valid_po = train_test_split(X_split_again_en, y_split_again_po, test_size=0.50, random_state=42)
```

```python
print("X_train: ", len(X_train_en))
print("y_train: ", len(y_train_po))
print("X_test: ", len(X_test_en))
print("y_test: ", len(y_test_po))
print("X_valid: ", len(X_valid_en))
print("y_valid: ", len(y_valid_po))
```
